```python
from pytesseract import Output
```

```python
img_manuscrito1 = cv2.imread("/content/teste_manuscrito_01.jpg")
img_manuscrito1 = cv2.cvtColor(img_manuscrito1, cv2.COLOR_BGR2RGB)
cv2_imshow(img_manuscrito1)
```

```python
def retornar_resultado(img, config):
  resultado = pytesseract.image_to_data(img, lang="por", config=config, output_type=Output.DICT)
  return resultado
```

```python
config = "--tessdata-dir tessdata"
resultado = retornar_resultado(img_manuscrito1, config)
resultado
```

```python
min_conf = 40

def caixa_texto(resultado, img, contador, cor=(255, 100, 0)):
  x = int(resultado["left"][contador])
  y = int(resultado["top"][contador])
  h = int(resultado["height"][contador])
  w = int(resultado["width"][contador])
  
  cv2.rectangle(img, (x, y), (x+w, y+h), cor, 2)
  
  return x, y, img

img_copia = img_manuscrito1.copy()

def desenhar_caixa(img, resultado, min_conf):
  for i in range(0, len(resultado["text"])):
    confianca = int(resultado["conf"][i])

    if confianca > min_conf:
      # Adicionando o texto que está escrito
      x, y, img = caixa_texto(resultado, img_copia, i)
      texto = resultado["text"][i]
      cv2.putText(img_copia, texto, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255))

desenhar_caixa(img_copia, resultado=resultado, min_conf=min_conf)
cv2_imshow(img_copia)
```

```python
# O putText do cv2 não suporte para acentuação
#Possível solução: Utilizar a biblioteca PIL
from PIL import ImageFont, Image, ImageDraw

caminho_font = "/content/calibri.ttf"

def escrever_texto(texto, x, y, img, font, tamanho_texto=33):
  #carregar fonte:
  fonte = ImageFont.truetype(font, tamanho_texto)
  #Lendo a imagem - vinda de um array
  img_pil = Image.fromarray(img)
  img_draw = ImageDraw.Draw(img_pil)
  img_draw.text((x, y - tamanho_texto), texto, font=fonte)
  img = np.array(img_pil)
  return img

def calcular_coordenadas(resultado, contador):
  x = int(resultado["left"][contador])
  y = int(resultado["top"][contador])
  h = int(resultado["height"][contador])
  w = int(resultado["width"][contador])

  return x, y, h, w

def desenhar_caixa(img, resultado, min_conf, cor=(255, 0, 0)):
  for i in range(0, len(resultado["text"])):
    confianca = int(resultado["conf"][i])

    if confianca > min_conf:
      # Adicionando o texto que está escrito
      x, y, h, w = calcular_coordenadas(resultado, i)
      
      texto = resultado["text"][i]
      img = escrever_texto(texto, x, y, img, caminho_font)
      
      # Desenhar Retangulo
      cv2.rectangle(img, (x, y), (x+w, y+h), cor, 2)
      
  return img
      # cv2.putText(img_copia, texto, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255))

img_copia = img_teste2.copy()
img_copia = desenhar_caixa(img_copia, resultado2, min_conf)
cv2_imshow(img_copia)
```

## Encontrando elementos específicos nas imagens
---

```python
import re # regular expressions
```

```python
img_tabela = cv2.imread("/content/tabela_teste.jpg")
img_tabela = cv2.cvtColor(img_tabela, cv2.COLOR_BGR2RGB)
cv2_imshow(img_tabela)
```

```python
config = "--tessdata-dir tessdata"
resultado_tabela = retornar_resultado(img_tabela, config)
resultado_tabela
```

```python
# Para identificar apenas as datas, vou precisar alterar a função desenhar_caixa

def desenhar_caixa(img, resultado, min_conf, regex, tamanho_fonte=28, cor=(255, 0, 0)):
  for i in range(0, len(resultado["text"])):
    confianca = int(resultado["conf"][i])

    if confianca > min_conf:
      # Adicionando o texto que está escrito
      texto = resultado["text"][i]
      if re.match(regex, texto):
        x, y, h, w = calcular_coordenadas(resultado, i)
        img = escrever_texto(texto, x, y, img, caminho_font, tamanho_fonte)

        # Desenhar Retangulo
        cv2.rectangle(img, (x, y), (x+w, y+h), cor, 2)
      
  return img

regex = "^(0?[1-9]|[12][0-9]|3[01])[\/\-](0?[1-9]|1[012])[\/\-]\d{4}$"
img_datas = desenhar_caixa(img_tabela.copy(), resultado_tabela, min_conf, regex, 16)
cv2_imshow(img_datas)
```
